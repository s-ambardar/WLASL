```python
import numpy as np
from tensorflow import keras
import pickle as pkl
```

```python
full_data = pkl.load(open('ASL_FINAL_data.p', 'rb'))
full_data_labels = pkl.load(open('ASL_FINAL_labels.p', 'rb'))
```

```python
label_is_00flat = full_data_labels[:,0] == 1
label_is_1D = full_data_labels[:,1] == 1
label_is_1X = full_data_labels[:,2] == 1
label_is_10A = full_data_labels[:,3] == 1
label_is_1515close = full_data_labels[:,4] == 1
label_is_2 = full_data_labels[:,5] == 1
label_is_20GL = full_data_labels[:,6] == 1
label_is_3PK = full_data_labels[:,7] == 1
label_is_5 = full_data_labels[:,8] == 1
label_is_5clawC = full_data_labels[:,9] == 1
label_is_88open = full_data_labels[:,10] == 1
label_is_9 = full_data_labels[:,11] == 1
label_is_None = full_data_labels[:,12] == 1
label_is_S = full_data_labels[:,13] == 1
```

```python
first_00flat_index = np.argmax(label_is_00flat)
first_1D_index = np.argmax(label_is_1D)
first_1X_index = np.argmax(label_is_1X)
first_10A_index = np.argmax(label_is_10A)
first_1515close_index = np.argmax(label_is_1515close)
first_2_index = np.argmax(label_is_2)
first_20GL_index = np.argmax(label_is_20GL)
first_3PK_index = np.argmax(label_is_3PK)
first_5_index = np.argmax(label_is_5)
first_5clawC_index = np.argmax(label_is_5clawC)
first_88open_index = np.argmax(label_is_88open)
first_9_index = np.argmax(label_is_9)
first_None_index = np.argmax(label_is_None)
first_S_index = np.argmax(label_is_S)
```

```python
indices = [
    first_00flat_index, 
    first_1D_index,
    first_1X_index,
    first_10A_index,
    first_1515close_index,
    first_2_index,
    first_20GL_index,
    first_3PK_index,
    first_5_index,
    first_5clawC_index,
    first_88open_index,
    first_9_index,
    first_None_index,
    first_S_index
]
for i in indices:
    print(i)
for i in indices:
    print(full_data_labels[i])
```

```python
import matplotlib.pyplot as plt

classes = ["0,0-flat", "1,D", "1-X", "10,A", "15,15-close", "2", "20,G,L", "3-P,K", "5", "5-claw,C", "8,8-open", "9", "None", "S"]

for k in range(len(indices)):
    index = indices[k]
    label = classes[k]
    top_row = [np.concatenate((full_data[index][0][i], full_data[index][1][i], full_data[index][2][i]),axis=0) for i in range(256)]
    middle_row = [np.concatenate((full_data[index][3][i], full_data[index][4][i], full_data[index][5][i]),axis=0) for i in range(256)]
    bottom_row = [np.concatenate((full_data[index][6][i], full_data[index][7][i], full_data[index][8][i]),axis=0) for i in range(256)]
    ninebynine = np.concatenate((top_row, middle_row, bottom_row),axis=0)
    plt.imshow(ninebynine)
    plt.show()
    plt.imsave(label+"_9x9.png", ninebynine)
    top_row = [np.concatenate((full_data[index][2][i], full_data[index][3][i]),axis=0) for i in range(256)]
    bottom_row = [np.concatenate((full_data[index][4][i], full_data[index][5][i]),axis=0) for i in range(256)]
    fourbyfour = np.concatenate((top_row, bottom_row),axis=0)
    plt.imshow(fourbyfour)
    plt.show()
    plt.imsave(label+"_4x4.png", fourbyfour)
```
