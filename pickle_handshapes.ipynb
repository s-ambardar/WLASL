```
import urllib.request

url_stub = 'https://www.handspeak.com/word/asl-eng/index.php?id='
base = "class=\"abc-btn\">Handshape"
end = "</a></li>"

def get_entry(num):
  file = urllib.request.urlopen('https://www.handspeak.com/word/asl-eng/index.php?id=' + str(num))
  for line in file:
    line = line.decode('utf-8')
    if base in line:
      i = line.index(base)
      j = line.index(end)
      hand = line[i+len(base)+1:j]
    if "How to sign" in line:
      i = line.index("How to sign ") + len("How to sign ")
      j = line.index(" in ASL")
      name = line[i:j]
  return name, hand
```

```
for i in range(5000,5050):
  n,h = get_entry(i)
  print(n,'\t\t', h)
```
